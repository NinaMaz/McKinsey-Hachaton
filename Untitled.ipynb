ride_type_desc – private or business order attribute (VARCHAR)

driver_response – driver choice of whether to accept the offer or not (VARCHAR)

# Feature Creation

```python
mean_accept = df.groupby(["driver_gk"])["driver_response"].mean()
mean_accept = mean_accept.rename(columns = {"driver_response":"mean_accept"})
```

```python
df_with_mean = pd.concat([df,mean_accept],join = "inner", axis = 1)
df_with_mean = df_with_mean.rename(columns = {df_with_mean.columns[-1]:"mean_accept"})
```

# Test/Validation

```python
def partition_users(l, percent):
    howManyNumbers = int(round(percent*len(l)/100))
    shuffled = l[ : ]
    random.shuffle(shuffled)
    return shuffled[howManyNumbers: ], shuffled[:howManyNumbers]

offer_list = list(np.unique(df["offer_gk"]))
train_offer_list,test_offer_list = partition_users(offer_list,0)
df_offer_train = df_with_mean[df_with_mean["offer_gk"].isin(train_offer_list)]
df_offer_test = df_with_mean[df_with_mean["offer_gk"].isin(test_offer_list)]
```

```python
df_offer_train = df_offer_train[interesting_values]
df_offer_test = df_offer_test[interesting_values]
```

## Dummies for categorical features

```python
df_offer_train = pd.concat([df_offer_train,pd.get_dummies(df_offer_train["offer_class_group"], prefix = "offer_class")],
                           join = "inner", axis = 1)
df_offer_train = df_offer_train.drop("offer_class_group",axis = 1)

df_offer_test = pd.concat([df_offer_test,pd.get_dummies(df_offer_test["offer_class_group"], prefix = "offer_class")],
                           join = "inner", axis = 1)
df_offer_test = df_offer_test.drop("offer_class_group",axis = 1)
```

```python
df_offer_train = pd.concat([df_offer_train,pd.get_dummies(df_offer_train["ride_type_desc"], prefix = "ride_type")],
                           join = "inner", axis = 1)
df_offer_train = df_offer_train.drop("ride_type_desc",axis = 1)

df_offer_test = pd.concat([df_offer_test,pd.get_dummies(df_offer_test["ride_type_desc"], prefix = "ride_type")],
                           join = "inner", axis = 1)
df_offer_test = df_offer_test.drop("ride_type_desc",axis = 1)
```

```python
df_offer_train = pd.concat([df_offer_train,pd.get_dummies(df_offer_train["hour_key"], prefix = "hour")],
                           join = "inner", axis = 1)
df_offer_train = df_offer_train.drop("hour_key",axis = 1)

df_offer_test = pd.concat([df_offer_test,pd.get_dummies(df_offer_test["hour_key"], prefix = "hour")],
                           join = "inner", axis = 1)
df_offer_test = df_offer_test.drop("hour_key",axis = 1)
```

```python
df_offer_train = pd.concat([df_offer_train,pd.get_dummies(df_offer_train["weekday_key"], prefix = "weekday")],
                           join = "inner", axis = 1)
df_offer_train = df_offer_train.drop("weekday_key",axis = 1)

df_offer_test = pd.concat([df_offer_test,pd.get_dummies(df_offer_test["weekday_key"], prefix = "weekday_key")],
                           join = "inner", axis = 1)
df_offer_test = df_offer_test.drop("weekday_key",axis = 1)
```

## Replacing -1 with 0 

```python
df_offer_train = df_offer_train.replace(-1,0)
df_offer_test = df_offer_test.replace(-1,0)
```

```python
X_train = df_offer_train.drop("driver_response",axis = 1).values
Y_train = df_offer_train["driver_response"].values

X_test = df_offer_test.drop("driver_response",axis = 1).values
Y_test = df_offer_test["driver_response"].values
```

## Without -1

```python
df_offer_train = df_offer_train[df["distance_km"]!=-1]
df_offer_train = df_offer_train[df["duration_min"]!=-1]

df_offer_test = df_offer_test[df["distance_km"]!=-1]
df_offer_test = df_offer_test[df["duration_min"]!=-1]
```

```python
X_train = df_offer_train.drop("driver_response",axis = 1).values
Y_train = df_offer_train["driver_response"].values

X_test = df_offer_test.drop("driver_response",axis = 1).values
Y_test = df_offer_test["driver_response"].values
```
